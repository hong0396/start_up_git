```python
import datetime
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller as ADF
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
```

```python
ts_simu200= pd.read_csv('data2.csv',index_col='t')
dates=pd.date_range(start='2017/01/01', periods=200)
ts_simu200.set_index(dates, inplace=True)
dta=ts_simu200['ARIMA']
```

```python
ts_simu200.head()
```

```python
ts_simu200.plot(figsize=(20,10))
plt.show()
```

### ARIMA建模分5步
####     1）探索平稳性

```python
#平稳性检验
result = ADF(dta)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

#### P值0.33 无法拒绝平稳假设 故该数据是非平稳 进行差分

```python
diff1= dta.diff(1)
diff1.dropna(inplace=True)
```

#### 2）绘制ACF与PACF 以定阶

```python
plot_acf(diff1,lags=100)
plot_pacf(diff1,lags=100)
plt.show()
```

#### ACF拖尾 PACF为1阶截尾 判断ARIMA（1，1，0）

#### ACF与PACF截尾 存在估计误差
#### 可使用AIC准则来识别 统计量越小越好

```python
(p, q) =(sm.tsa.arma_order_select_ic(diff1,max_ar=3 ,max_ma=3 ,ic='aic')['aic_min_order'])
print(p,q)
```

#### 这里给出p=1，q=0 与ACF和PACF方法一致
#### 有些不太好的数据（专指非学术派数据 非严格周期数据 指实际工业环境产生的数据化） 两种方法给出的参数会不同

#### 至于这两种判别p、q参数的方法 没有严格意义的谁好谁差
#### 实测了一些数据 有的ACF和PACF判别法好一些 有的则AIC判别给出的参数更合理
#### 两者方法均可测试

#### 3）ARIMA模型建模

```python
arima110 = sm.tsa.ARIMA(dta,(1,1,0)).fit()
```

#### 4）残差白噪声检验

```python
resid = arima110.resid #残差
fig = plt.figure(figsize=(12,8))
plot_acf(resid.values.squeeze(), lags=100)
plot_pacf(resid, lags=100)
plt.show()
```

#### 5）预测

```python
# fig = arima110.plot_predict('20170102',
#                         '20170820', 
#                             dynamic=False, plot_insample=True)
fig = arima110.plot_predict(pd.to_datetime('2017-01-02'),
                        pd.to_datetime('2017-01-01') + datetime.timedelta(days = len(ts_simu200)+10), 
                            dynamic=False, plot_insample=True)
fig.set_figwidth(15)
fig.set_figheight(8)
plt.show()
```

#### ARIMA也有自动调参的包
####  pip install pyramid-arima 
####  这包可以自动调参
####  预测的效果嘛 也一般 和手动识别参数差不了多少 
####  有时候与手动调参的参数差很多
####  就是能省点力 不用画ACF和PACF

```python
data = np.array(ts_simu200.values).T[0]
```

```python
from pyramid.arima import auto_arima
# 导入包。

fittedmodel = auto_arima(data, start_p=1, start_q=1, max_p=9, max_q=6, max_d=3,max_order=None,
                         seasonal=False, m=1, test='adf', trace=False,
                         error_action='ignore',  # don't want to know if an order does not work
                         suppress_warnings=True,  # don't want convergence warnings
                         stepwise=True, information_criterion='bic', njob=-1)  # set to stepwise
# 把训练数据放入auto_arima得到最优模型
# ARIMA里的三个参数PDQ都是可以进行自动调参的，就是通过调整start_p和max_p
# 它会自动对这三个参数调整，这里m=1的意思是不考虑季节性
# 考虑季节性 m=4 一季周期 m=12 一年周期
fittedmodel.summary()
```

#### 自动调参给出的模型是ARMA（2,0）

```python
#导入画图包
from bokeh.plotting import figure, show, output_notebook
import pandas as pd

# init bokeh
output_notebook()

def plot_arima(truth, forecasts, title="ARIMA", xaxis_label='Time',
               yaxis_label='Value', c1='#A6CEE3', c2='#B2DF8A', 
               forecast_start=None, **kwargs):
    
    # make truth and forecasts into pandas series
    n_truth = truth.shape[0]
    n_forecasts = forecasts.shape[0]
    
    # always plot truth the same
    truth = pd.Series(truth, index=np.arange(truth.shape[0]))
    
    # if no defined forecast start, start at the end
    if forecast_start is None:
        idx = np.arange(n_truth, n_truth + n_forecasts)
    else:
        idx = np.arange(forecast_start, n_forecasts)
    forecasts = pd.Series(forecasts, index=idx)
    
    # set up the plot
    p = figure(title=title, plot_height=400, **kwargs)
    p.grid.grid_line_alpha=0.3
    p.xaxis.axis_label = xaxis_label
    p.yaxis.axis_label = yaxis_label
    
    # add the lines
    p.line(truth.index, truth.values, color=c1, legend='Observed')
    p.line(forecasts.index, forecasts.values, color=c2, legend='Forecasted')
    
    return p
y_hat = np.append(fittedmodel.predict_in_sample(),fittedmodel.predict(10))
# y_pre = fittedmodel.predict
show(plot_arima(data, y_hat, 
                title="Original Series & In-sample Predictions", 
                c2='#FF0000', forecast_start=0))
```
